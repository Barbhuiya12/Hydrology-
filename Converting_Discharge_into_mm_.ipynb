```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_excel("Patan.xlsx")
df 
```

**Function which is going to convert Q in cum to Q mm**

```python
A= float(input("Enter the area of the Catchment :"))
def convert(Q):
     
    return ((Q * 1000 * 86400) / A)
```

**Applyin Above formule in data above dataframe**

```python
df["Qmm"]=df["Discharge"].apply(lambda Q: convert(Q))
df
```
