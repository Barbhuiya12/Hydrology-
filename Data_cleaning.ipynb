<h1>Missing Value 

```python
create_nan={"Examples":[np.nan,np.nan,20,40,5,0,6,np.nan,30,50,70]}
df_nan=pd.DataFrame(create_nan,columns={'Examples'})
df_nan
```

Check the nan in the data set 

```python
df_nan.isnull().values.any()
##Checking specific column 
df_nan["Examples"].isnull().values.any()
```

Fill the nan with zeo 


```python
df_clean=df_nan.fillna(0)
```

Filling nan with backward value (Forward(ffill) or backword(bfill) propagation)


```python
df_nan.fillna(method='ffill')
```

Droping the missing data 

```python
df_drop=df_nan.dropna()
df_drop.describe()
```
